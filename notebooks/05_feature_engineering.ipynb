# Notebook 5: Feature Engineering (Vectorization)
In this notebook, we will load our preprocessed text data, use our modular `src/features.py` package to apply **TF-IDF Vectorization** (Term Frequency - Inverse Document Frequency), and explore the resulting numerical features.

```python
import pandas as pd
import os
from src.features import get_tfidf_vectorizer

# Load processed dataset
data_path = os.path.join("..", "data", "processed_jobs.csv")
df = pd.read_csv(data_path)

# Handle any null strings (in case clean_text became empty/nan)
df['clean_text'] = df['clean_text'].fillna('')

print("Shape of processed data:", df.shape)
```

## 1. Apply TF-IDF Vectorization
We will initialize our TF-IDF Vectorizer with `max_features=5000` (keeping top 5,000 most predictive words/n-grams) and extract the text features.

```python
vectorizer = get_tfidf_vectorizer(max_features=5000, ngram_range=(1, 2))

# Fit and transform clean text
X_tfidf = vectorizer.fit_transform(df['clean_text'])

print("Type of vectorized matrix:", type(X_tfidf))
print("Shape of TF-IDF matrix:", X_tfidf.shape)
```

## 2. Inspecting the Vocabulary
Let's print some of the terms that our vectorizer has captured.

```python
vocab = vectorizer.get_feature_names_out()
print("Total vocabulary size:", len(vocab))
print("Sample features (every 250th word):\n", vocab[::250])
```
