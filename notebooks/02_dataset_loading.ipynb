# Notebook 2: Dataset Loading and Initial Inspection
In this notebook, we will load our `fake_job_postings.csv` dataset using **Pandas** and explore its structure.

```python
import pandas as pd
import os

# Define the path to our dataset
data_path = os.path.join("..", "data", "fake_job_postings.csv")

# Load the dataset
df = pd.read_csv(data_path)

# Show the first 5 rows
df.head()
```

## 1. Introduction to NumPy: Analyzing Array Dimensions and Target Classes
Under the hood, Pandas stores its data columns as **NumPy arrays**. Let's convert our target column (`fraudulent`) into a raw NumPy array and inspect its shape and distribution using NumPy functions.

```python
import numpy as np

# Convert the 'fraudulent' column into a NumPy array
target_array = df['fraudulent'].values

print("Type of target_array:", type(target_array))
print("Shape of target_array (dimensions):", target_array.shape)
print("Total number of items:", len(target_array))

# Count occurrences of 0 (real) and 1 (fake) using NumPy
counts = np.bincount(target_array)
print("\nLegitimate jobs (Class 0):", counts[0])
print("Fraudulent jobs (Class 1):", counts[1])
print(f"Percentage of fraudulent jobs: {counts[1] / len(target_array) * 100:.2f}%")
```

## 2. Introduction to Matplotlib & Seaborn: Visualizing Class Imbalance
Data visualization is crucial. Let's use **Matplotlib** and **Seaborn** to create a bar plot showing the ratio of legitimate to fraudulent jobs.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a premium visual style
sns.set_theme(style="whitegrid")
plt.figure(figsize=(6, 5))

# Create a bar plot using Seaborn
ax = sns.countplot(x='fraudulent', data=df, palette="viridis")

# Add labels and title
plt.title("Distribution of Real vs. Fake Job Postings", fontsize=14, pad=15)
plt.xlabel("Class (0 = Legitimate, 1 = Fraudulent)", fontsize=12)
plt.ylabel("Number of Job Postings", fontsize=12)
ax.set_xticklabels(["Legitimate (0)", "Fraudulent (1)"])

# Show the plot
plt.show()
```
